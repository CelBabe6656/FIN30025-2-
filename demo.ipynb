# TradieTax AI Core Logic - Google Colab / Python Demo
This notebook demonstrates the core tax calculations and audit risk logic used in the TradieTax application, without the React frontend.

```python
from tax_engine import TradieTaxEngine

# 1. Initialize the Engine
engine = TradieTaxEngine(user_category='Sole Trader', is_gst_registered=False)

# 2. Add some data
engine.add_income(5000, '2026-04-01', 'Sales', 'Kitchen Cabinets Install')
engine.add_income(12000, '2026-04-10', 'Sales', 'Full Bathroom Renovation')

engine.add_receipt('Bunnings', 450.75, 'Materials')
engine.add_receipt('Tradelink', 1200.00, 'Plumbing Parts', is_asset=True)
engine.add_receipt('OfficeWorks', 150.00, 'Office') # Error: Round number detected

# 3. Run Calculations
totals = engine.calculate_totals()
tax = engine.estimate_tax(totals['taxable_income'])
risks = engine.run_audit()

# 4. Display Results
print(f"Total Turnover: ${totals['turnover']:,.2f}")
print(f"Taxable Profit: ${totals['net_profit']:,.2f}")
print(f"Estimated Income Tax: ${tax:,.2f}")
print("\n--- Audit Report ---")
for risk in risks:
    print(f"{risk['risk']}: {risk['issue']} - {risk['advice']}")
```
